## 1. Rolling Mean
In statistics, a moving average (rolling average or running average or moving mean[1] or rolling mean) is a calculation to analyze data points by creating a series of averages of different selections of the full data set. Variations include: simple, cumulative, or weighted forms.

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(euro_to_dollar["Time"],euro_to_dollar["US_dollar"])
plt.show()
```

If we look at the line's shape, we see many small wiggles — rather than seeing a smooth line. The wiggles, however, have meaning: : they are the visual representation of the daily variation in the exchange rate. The rate goes up and down, up and down again, day to day. The rate only shows clear upward or downward trends in the longer run (months or years).

Depending on our goals, we may not want to show that daily variation on our graph. If we want to hide it and show only the long-term trends, we can use the rolling mean (also known as the moving average).

```python
values=pd.DataFrame()
```

```python
pd.Series(range(1,20,2))
```

```python
values['daily_values']=pd.Series(range(1,20,2))
values
```

```python
values["rolling_mean_2"]=values["daily_values"].rolling(2).mean() # it will add 2 values and then take average
values
```

```python
values["rolling_mean_3"]=values["daily_values"].rolling(3).mean()
values
```

```python
values["rolling_mean_5"]=values["daily_values"].rolling(5).mean()
values
```

```python
plt.figure(figsize=(9,6))

plt.subplot(3,2,1)
plt.ylim(bottom=0.8,top=1.65)
plt.plot(euro_to_dollar["Time"],euro_to_dollar["US_dollar"])
plt.title("Original Values",weight='bold')

for i,rolling_mean in zip([2,3,4,5,6],[7,30,50,100,365]):
    plt.subplot(3,2,i)
    plt.ylim(bottom=0.8,top=1.65)
    plt.plot(euro_to_dollar["Time"],euro_to_dollar["US_dollar"].rolling(rolling_mean).mean())
    plt.title(f"Rolling Window: {rolling_mean}",weight='bold')

plt.tight_layout() # Adjust the padding between and around subplots.
plt.show()
```

## `Financial Crisis 2007-2008`

```python
financial_crisis=euro_to_dollar.copy()[(euro_to_dollar["Time"].dt.year>=2006) & (euro_to_dollar["Time"].dt.year<=2009)]
financial_crisis
```

```python
financial_crisis_7_8=euro_to_dollar.copy()[(euro_to_dollar["Time"].dt.year>=2007) & (euro_to_dollar["Time"].dt.year<=2008)]
financial_crisis_7_8
```

```python
import matplotlib.style as style
style.use('fivethirtyeight')

# Adding the plot
fig,ax=plt.subplots(figsize=(8,3))
ax.plot(financial_crisis["Time"],financial_crisis["rolling_mean"],linewidth=1,color='gray')

# Highlighting 2007-2008 Crisis
ax.plot(financial_crisis_7_8["Time"],financial_crisis_7_8["rolling_mean"],linewidth=2,color='red')

# ax.set_xticklabels(["2006","","2007","","2008","","2009","","2010"],alpha=0.5)

ax.set_xticklabels([])

x=0.018
for year in ["2006","2007","2008","2009","2010"]:
    ax.text(x,-0.08,year,alpha=0.5,fontsize=11,transform=plt.gca().transAxes)
    x += 0.23

ax.set_yticklabels([])

y=0.07
for rate in ["1.2","1.3","1.4","1.5"]:
    ax.text(-0.04,y,rate,alpha=0.5,fontsize=11,transform=plt.gca().transAxes)
    y += 0.235
    
# Adding a title and a sub-title
ax.text(-0.05,1.2,"Euro-USD rate peaked to 1.59 during 2007-2008 financial crisis",
        weight="bold",transform=plt.gca().transAxes)
ax.text(-0.05,1.1,"Euro-USD exchange rates between 2006 and 2010",
        size=12,transform=plt.gca().transAxes)

# Adding a signature
ax.text(-0.05,-0.25,"TECHMA ZONE" + " "*80 + "Source: European Central Bank",color="#f0f0f0",backgroundcolor="#4d4d4d",
        size=12,transform=plt.gca().transAxes)

ax.axvspan(xmin=pd.to_datetime("2008-03-15"),xmax=pd.to_datetime("2008-09-10"),ymin=0.085,ymax=1,alpha=0.4,color="grey")

plt.show()
```

## `The Four US-Presidencies Comparison`

```python
bush_obama_trump_biden= euro_to_dollar.copy(
)[(euro_to_dollar["Time"].dt.year >= 2001) & (euro_to_dollar["Time"].dt.year <= 2024)]

bush = bush_obama_trump_biden.copy()[bush_obama_trump_biden["Time"].dt.year <= 2008]

obama = bush_obama_trump_biden.copy()[(bush_obama_trump_biden["Time"].dt.year >= 2009)
                                      & (bush_obama_trump_biden["Time"].dt.year <= 2016)]

trump = bush_obama_trump_biden.copy()[(bush_obama_trump_biden["Time"].dt.year >= 2017)
                                      & (bush_obama_trump_biden["Time"].dt.year <= 2020 )]

biden = bush_obama_trump_biden.copy()[bush_obama_trump_biden["Time"].dt.year >= 2021]
```

Below, you will notice we used matplotlib's functional approach to build the graphs. We use this approach because it offers more flexibility in arranging the subplots:
* We first build four of the graphs on a 2-by-4 grid (this grid should have eight subplots, but we only build four; the bottom row remains empty)
* We then build only the bottom graph on a 2-by-1 grid (this grid should have two subplots; the top row remains empty)
* The two grids are merged, and we end up with four graphs on the top row and one graph on the bottom row

```python
import matplotlib.style as style
style.use("fivethirtyeight")

# Adding the subplots
plt.figure(figsize=(14,8))

# pattern 1
ax1 = plt.subplot(2,4,1) # row, col, index
ax2 = plt.subplot(2,4,2)
ax3 = plt.subplot(2,4,3)
ax4 = plt.subplot(2,4,4)

# pattern 2
ax5 = plt.subplot(2,1,2) # row, col, index

axes=[ax1,ax2,ax3,ax4,ax5]

# Formatting all the subplots
for ax in axes:
    ax.set_ylim(bottom=0.8,top=1.7)
    ax.set_yticks([1.0,1.2,1.4,1.6])
    ax.set_yticklabels(['1.0','1.2','1.4','1.6 $'], alpha=0.4,size=12)
    
# Ax1: Bush
ax1.plot(bush["Time"],bush["rolling_mean"],linewidth=3,color='red')
ax1.set_xticklabels(["","2001","","2003","","2005","","2007","","2009"],
                    alpha=0.4, size=12)
ax1.text(0.08,2.32,"BUSH",fontsize=15, color='red',weight="bold",
         transform=plt.gca().transAxes)
ax1.text(0.06,2.25,"(2001-2008)",fontsize=12, weight="bold",
         transform=plt.gca().transAxes, alpha=0.5)

# Ax2: Obama
ax2.plot(obama["Time"],obama["rolling_mean"],linewidth=3,color='blue')
ax2.set_xticklabels(["","2009","","2011","","2013","","2015","","2017"],
                    alpha=0.4, size=12)
ax2.text(0.34,2.32,"OBAMA",fontsize=15, color='blue',weight="bold",
         transform=plt.gca().transAxes)
ax2.text(0.33,2.25,"(2009-2016)",fontsize=12, weight="bold",
         transform=plt.gca().transAxes, alpha=0.5)

# Ax3: Trump
ax3.plot(trump["Time"],trump["rolling_mean"],linewidth=3,color='green')
ax3.set_xticklabels(["2017","","2018","","2019","","2020","","2021"],
                    alpha=0.4, size=12)
ax3.text(0.6,2.32,"TRUMP",fontsize=15, color='green',weight="bold",
         transform=plt.gca().transAxes)
ax3.text(0.59,2.25,"(2017-2020)",fontsize=12, weight="bold",
         transform=plt.gca().transAxes, alpha=0.5)

# Ax4: Biden
ax4.plot(biden["Time"],biden["rolling_mean"],linewidth=3,color='purple')
ax4.set_xticklabels(["2021","","","2022","","","2023","","","2024"],
                    alpha=0.4, size=12)
ax4.text(0.85,2.32,"BIDEN",fontsize=15, color='purple',weight="bold",
         transform=plt.gca().transAxes)
ax4.text(0.82,2.25,"(2021-Present)",fontsize=12, weight="bold",
         transform=plt.gca().transAxes, alpha=0.5)

# Ax5: Combining All
ax5.plot(bush["Time"],bush["rolling_mean"],linewidth=3,color='red')
ax5.plot(obama["Time"],obama["rolling_mean"],linewidth=3,color='blue')
ax5.plot(trump["Time"],trump["rolling_mean"],linewidth=3,color='green')
ax5.plot(biden["Time"],biden["rolling_mean"],linewidth=3,color='purple')
ax5.set_xticks([])

# Adding a title and a sub-title
ax.text(-0.05,2.55,"Euro-USD rate averaged 1.188 under the last four US President's",
        weight="bold",transform=plt.gca().transAxes)
ax.text(-0.05,2.45,"Euro-USD exchange rates under Bush, Obama, Trump and Biden",
        size=12,transform=plt.gca().transAxes)

# Adding a signature
ax.text(-0.05,-0.15,"TECHMA ZONE" + " "*172 + "Source: European Central Bank",color="#f0f0f0",backgroundcolor="#4d4d4d",
        size=12,transform=plt.gca().transAxes)

plt.show()
```
